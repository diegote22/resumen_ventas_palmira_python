# Análisis de Ventas de un Local en Palmira

## Objetivo
El objetivo de este programa en Python es analizar un conjunto de datos de ventas de una tienda en Palmira. Utilizaremos técnicas de Data Science para extraer información valiosa sobre el rendimiento de las ventas.

## Consigna
Lectura de Datos:

Primero, crearemos un DataFrame utilizando la biblioteca Pandas. Este DataFrame se llenará con los datos del archivo Datos_Ventas_Tienda.csv. Este archivo contiene información crucial como la fecha de venta, la categoría del producto, la cantidad vendida y el precio.
Fusión de Datos:

Luego, cargaremos un segundo conjunto de datos desde el archivo Datos_Ventas_Tienda2.csv.
Después, concatenaremos ambos DataFrames en uno solo. Esto nos dará una vista completa de todos los datos de ventas en un solo lugar.
Tratamiento de Datos:

Ahora, usaremos Pandas para limpiar, filtrar y transformar los datos. Esto puede incluir eliminar entradas duplicadas, manejar valores faltantes y convertir tipos de datos si es necesario.
Análisis de Ventas:

Con los datos limpios, realizaremos un análisis para responder preguntas clave:
¿Cuál es el producto más vendido en general?
¿Cuál es el mes en el que se registraron más ventas?
Datos Agrupados:

Después, agruparemos los datos por categoría de producto. Esto nos permitirá analizar las ventas de cada categoría individualmente y ver cuáles son las más populares.
Guardar Resultados:

Finalmente, guardaremos el DataFrame completo (incluyendo la columna de meses que crearemos) en un nuevo archivo CSV en tu ordenador. Este archivo contendrá todos los datos procesados y los resultados de nuestro análisis.

```python
import pandas as pd
```

```python
ruta = "C:/Users/diego/Desktop/Local_Palmira/Datos_Ventas_Local_Palmira.csv"
df_ventas_palmira = pd.read_csv(ruta)
df_ventas_palmira
```

```python
ruta2 = "C:/Users/diego/Desktop/Local_Palmira/Datos_Ventas_Local2_Palmira.csv"
df_ventas_palmira2 = pd.read_csv(ruta2)
df_ventas_palmira2
```

```python
df = pd.concat([df_ventas_palmira, df_ventas_palmira2], ignore_index=True)
df.tail()
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
```

```python
df
```

```python
df.info()
```

# Producto más vendido

```python
producto_mas_vendido = df.groupby('Producto')['Cantidad'].sum()
producto_mas_vendido = producto_mas_vendido.sort_values(ascending=False)
producto_mas_vendido.head(1)
```

# ¿Cuál es el mes con más ventas?

```python
# Creamos una columna 'Meses'

meses = []
for f in df['Fecha']:
    meses.append(f.month)

df['Meses'] = meses
df
```

```python
# Ahora vamos a ver cual es el Mes con más ventas:

mes_con_mas_ventas = df.groupby('Meses')['Total Venta'].sum().sort_values(ascending=False)
mes_con_mas_ventas.head(1)
```

# Datos Agrupados:

```python
ventas_por_categorias = df.groupby('Producto')['Total Venta'].sum()
ventas_por_categorias
```
